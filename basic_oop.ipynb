```python
# Q:->explain class and object with respect oops ang give example
# In object-oriented programming, a 'class' is a blueprint or template that defines the properties (attributes) and behaviors (methods)
# that objects of that class will have. It's like a recipe or a set of instructions for creating something.

# An object, on the other hand, is an instance of a class. It's a concrete realization of the blueprint defined by the class.
# Think of it as a specific item created using the recipe.

# Imagine we want to represent a car in our code. we would create a class called "Car" that defines properties like color, model, and year, 
# as well as behaviors like start, stop, and accelerate.

class car:
    def __init__(self,color,model,year):
        self.color = color
        self.model = model
        self.year  = year

    def get_info(self):
        return self.color,self.model,self.year

my_fav_car = car ('white',2024,'fresh') 
my_fav_car.get_info()
        
```
